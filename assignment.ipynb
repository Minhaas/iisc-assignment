```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = pd.read_json("individuals.json")
data.head(10)
```

```python
ageDensity, _ = np.histogram(data["age"], bins=max(data["age"]), density="True")
print(ageDensity[:10])
```

```python
householdSize = np.bincount(data["household"])
householdDensity, _ = np.histogram(householdSize, bins=max(householdSize), density="True")
print(len(householdSize))
print(householdSize[:10])
print(householdDensity[:10])
```

```python
plt.figure("Age Density")
plt.plot(ageDensity, color='r', marker='o')
plt.grid()
plt.title("Age Distribution")
plt.xlabel("age")
plt.ylabel("density")
```

```python
plt.figure("Household density")
plt.plot(householdDensity, color='r', marker='o')
plt.grid()
plt.title("Household density")
plt.xlabel("Household size")
plt.ylabel("Density")
```
